```python
import whisper
import glob
import os

# Load Whisper model
model = whisper.load_model("base")

# Store results
whisper_results = []

audio_files = sorted(glob.glob("audio/*.wav"))

for i, audio_file in enumerate(audio_files):

    result = model.transcribe(audio_file)

    transcript = result["text"]

    record = {
        "segment_id": i,
        "start_time": i * 10,
        "end_time": (i + 1) * 10,
        "transcript": transcript
    }

    whisper_results.append(record)

    print(f"\nSegment {i}")
    print(transcript)

print("\nWhisper Transcription Complete!")
```

```python
from panns_inference import AudioTagging
import librosa
import glob
import numpy as np

audio_tagging = AudioTagging(device='cpu')

panns_results = [] # Initialize panns_results as an empty list

for i, audio_file in enumerate(sorted(glob.glob("audio/*.wav"))):

    waveform, sr = librosa.load(
        audio_file,
        sr=32000,
        mono=True
    )

    clipwise_output, embedding = audio_tagging.inference(
        waveform[None, :]
    )

    top_indices = np.argsort(
        clipwise_output[0]
    )[::-1][:5]

    sound_events_list = []
    for idx in top_indices:
        sound_events_list.append({
            "event": audio_tagging.labels[idx],
            "score": float(clipwise_output[0][idx])
        })

    panns_results.append({
        "segment_id": i,
        "start_time": i * 10, # Assuming 10-second segments
        "end_time": (i + 1) * 10, # Assuming 10-second segments
        "sound_events": sound_events_list
    })

    print(f"\nSegment {i}: {audio_file}")
    for event in sound_events_list:
        print(f"{event['event']} {round(event['score'], 3)}")

print("\nPANNs Audio Analysis Complete!")
```

```python
import os
import glob
import json
import base64

vlm_results = []

frame_folders = sorted(glob.glob("frames/*"))

for i, folder in enumerate(frame_folders):

    frame_paths = sorted(glob.glob(f"{folder}/*.jpg"))

    # Use first 3 frames
    frame_paths = frame_paths[:3]

    prompt = """
Analyze these video frames.

Return ONLY a valid JSON object in this format:

{
    "people_count": number,
    "objects": [],
    "actions": [],
    "scene_summary": ""
}
"""

    # Build multimodal message
    content = [
        {
            "type": "text",
            "text": prompt
        }
    ]

    # Add every frame
    for frame in frame_paths:

        with open(frame, "rb") as img:
            image_data = base64.b64encode(img.read()).decode("utf-8")

        content.append(
            {
                "type": "image_url",
                "image_url": {
                    "url": f"data:image/jpeg;base64,{image_data}"
                }
            }
        )

    # Call Groq Vision
    response = groq_client.chat.completions.create(
        model="meta-llama/llama-4-scout-17b-16e-instruct",
        messages=[
            {
                "role": "user",
                "content": content
            }
        ],
        temperature=0
    )

    answer = response.choices[0].message.content

    record = {
        "segment_id": i,
        "start_time": i * 10,
        "end_time": (i + 1) * 10,
        "vlm_analysis": answer
    }

    vlm_results.append(record)

    print("=" * 60)
    print(f"Segment {i}")
    print(answer)

print("\nGroq Vision Analysis Complete!")
```

```python
import json

unified_records = []

for i in range(len(whisper_results)):

    record = {
        "segment_id": i,
        "start_time": whisper_results[i]["start_time"],
        "end_time": whisper_results[i]["end_time"],
        "speech_transcript": whisper_results[i]["transcript"],
        "sound_events": panns_results[i]["sound_events"],
        "visual_analysis": vlm_results[i]["vlm_analysis"]
    }

    unified_records.append(record)

print("Total Records:", len(unified_records))

with open("unified_records.json", "w") as f:
    json.dump(unified_records, f, indent=4)
```

```python
from sentence_transformers import SentenceTransformer
import json
import numpy as np

# Load embedding model
embedding_model = SentenceTransformer("all-MiniLM-L6-v2")

documents = []

for record in unified_records:

    # -----------------------------
    # Convert sound events to text
    # -----------------------------
    sound_text = ", ".join(
        [event["event"] for event in record["sound_events"]]
    )

    # -----------------------------
    # Convert visual analysis to text
    # -----------------------------
    visual = record["visual_analysis"]

    if isinstance(visual, dict):

        visual_text = f"""
People Count: {visual.get('people_count', 'Unknown')}

Objects:
{", ".join(visual.get("objects", []))}

Actions:
{", ".join(visual.get("actions", []))}

Scene Summary:
{visual.get("scene_summary", "")}
"""

    else:
        visual_text = str(visual)

    # -----------------------------
    # Create searchable document
    # -----------------------------
    text = f"""
Time: {record['start_time']} - {record['end_time']} seconds

Speech Transcript:
{record['speech_transcript']}

Sound Events:
{sound_text}

Visual Analysis:
{visual_text}
"""

    documents.append(text)

# -----------------------------
# Generate Embeddings
# -----------------------------
embeddings = embedding_model.encode(
    documents,
    convert_to_numpy=True,
    show_progress_bar=True
)

print("Total Documents :", len(documents))
print("Embedding Shape :", embeddings.shape)

# -----------------------------
# Save Embeddings
# -----------------------------
np.save("embeddings.npy", embeddings)

# Save Documents
with open("documents.json", "w") as f:
    json.dump(documents, f, indent=4)

print("\nEmbeddings Generated Successfully!")
print("Files Saved:")
print("1. embeddings.npy")
print("2. documents.json")
```

```python
import chromadb

DB_PATH = "/content/drive/MyDrive/VideoRAG/video_rag_db"

chroma_client = chromadb.PersistentClient(path=DB_PATH)
# Delete previous collection if it exists
try:
    chroma_client.delete_collection("video_segments")
except:
    pass

collection = chroma_client.create_collection(
    name="video_segments"
)

for i, (doc, emb) in enumerate(zip(documents, embeddings)):

    collection.add(
        ids=[f"segment_{i}"],
        documents=[doc],
        embeddings=[emb.tolist()],
        metadatas=[{
            "segment_id": i,
            "start_time": unified_records[i]["start_time"],
            "end_time": unified_records[i]["end_time"]
        }]
    )

print(f"Stored {collection.count()} segments in ChromaDB.")
```

```python
# ==========================================
# Video RAG Question Answer Function
# ==========================================

def ask_video(question, top_k=3):

    query_embedding = embedding_model.encode(
        question,
        convert_to_numpy=True
    )

    results = collection.query(
        query_embeddings=[query_embedding.tolist()],
        n_results=top_k
    )

    context = "\n\n".join(results["documents"][0])

    response = groq_client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[
            {
                "role": "user",
                "content": f"""
Context:
{context}

Question:
{question}

Answer:
"""
            }
        ],
        temperature=0
    )

    return response.choices[0].message.content
```
